# መግቢያ

ባለፈው ክፍል መሰረታዊ የሆኑትን የፓንዳስ ውሂብ አይነቶችን አይተናል። ሲሪስ እና ዳታ ፍሬም። በዚህ ክፍል ሎክ በመጠቀም ውሂቡን እንዴት መተንተን እንደምንችል እንመለከታለን። 

# ሎክ (loc)

```python
# አስፈላጊውን ጥቅል ወደ ፕሮግራማችን እንጥራ 
import pandas as pd
```

```python
# ውሂቡን እናንብብ 
ዳታ_ፍሬም = pd.read_csv('./data/ufo.csv')
```

```python
# ውሂቡ ምን እንደሚመስል እንመልከት 
ዳታ_ፍሬም.head(5)
```

ሎክ በሁለት አይነት ዋና ዋና መንገዶች ውሂብ ለይቶ ለማውጣት ጥቅም ላይ ይውላል። 
* ምልክት በመጠቀም (ሌብል በመጠቀም) 
* ቡልያን በመጠቀም  (እውነት ውሸት አይነት ምርጫ በመጠቀም)

```python
# ምልክትን በመጠቀም ውሂብን ለይቶ ማውጣት 
# ዳታ_ፍሬም.loc[ረድፍ_ጠቋሚ, አምድ_ጠቋሚ]
# row indexer [ረድፍ_ጠቋሚ]
# column indexer [አምድ_ጠቋሚ]
ዳታ_ፍሬም.loc[0:4, 'City':'Time']
```

```python
# ሎክ ጨማሪ ወይም ኢንክሉሲቭ ነው። 
ዳታ_ፍሬም.loc[0:4]  
#ሎክን ሳይጠቀሙም ተመሳሳይ ውሂብ ማውጣት ይቻላል። 
#ዳታ_ፍሬም[0:5]
```

```python
# ሌላው የፈለግነውን አምድ ጠቋሚ በማድረግ ውሂብን መለየት ይቻላል። 
# ኒውዮርክ ግዛት የታዩትን ይፎዎች መለየት ብንፈልግ 
# 1 Stateን ማውጫ እናድርገው 
ዳታ_ፍሬም_01 = ዳታ_ፍሬም.set_index('State')
# 2 በመቀጠል ሎክን በመጠቀም መለየት እንስራ 
ዳታ_ፍሬም_01_NY= ዳታ_ፍሬም_01.loc['NY']
```

```python
# ዳታ_ፍሬም_01_NY ምን እንደሚመስል እንመልከት 
ዳታ_ፍሬም_01_NY.head(5)
```

```python
# ማውጫውን ወደነባሪው መመለስ 
ዳታ_ፍሬም_01_NY.reset_index(inplace=True)
ዳታ_ፍሬም_01_NY.head(5)
```

```python
# ሲቲ እና ታይምን ለይቶ ማውጣት ቢያስፈልግ 
# ሁለት ቅንፍ መኖሩን ያስተውሉ 
ዳታ_ፍሬም_01_NY[['City', 'Time']].head(5)
```

```python
# ተመሳሳይ አይነት ውሂብ ሎክን በመጠቀም ለይቶ ለማውጣት 
ዳታ_ፍሬም_01_NY.loc[:, ['City', 'Time']].head()
```

```python
# ቡሊያንን ለመጠቀም ውሂቡ ምን እንደሚመስል ለመመልከት 
ዳታ_ፍሬም.head()
```

```python
# ኒውዮርክ የሆኑትን ለይተን ሲቲ እና ታይምን እንዘርዝር 
# ቡሊያንን መጠቀም 
ዳታ_ፍሬም.loc[ዳታ_ፍሬም['State']=='NY', ['City', 'Time']].head()
```

```python
# ኒወርክ ውስጥ የታዩት የትሪአንግል አይነት ቅርጽ ያላቸውን ዩፎዎችን ቀለም ለይተን እናውጣ 
ዳታ_ፍሬም.loc[(ዳታ_ፍሬም['State']=='NY') & (ዳታ_ፍሬም['Shape Reported']=='TRIANGLE') 
           & (ዳታ_ፍሬም['Colors Reported'].notnull()), 
           ['City', 'Colors Reported', 'Time']]
```

በውጤቱ መሰረት በ ኒውዮርክ ግዛት ውስጥ የታዩት ዩፎዎች ቀለማቸው ቀይ፣ ብርቱካናማ፣ ሰማያዊ፣ ቢጫ አረንጓዴ እና የቀይ አረንጓዴ ናቸው ማለት ነው።
