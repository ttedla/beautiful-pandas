## ውሂብ ጎታ (ዳታቤዝን) ማንበብ

```python
# ስኪኤል ላይት ዳታቤዝ ለማንበብ 
# sqlite3 የፓይተን sqlite ውሂብ ጎታን ለማንበብ የሚረዳ ጥቅል ነው። 
import sqlite3

# sqlite3.connect በመጠቀም ከ ውሂብ ጎታው ጋር እንገናኛለን
መገናኛ = sqlite3.connect('./data/weather_2012.sqlite')

# መጠይቃችንን (ክዌሪያችንን) እናዘጋጃለን። 
# ስለውሂብ ጎታ ስማችን ወይም በውስጡ ስለሚገኙት ሰንጠረዦች ስም ማወቅ አለብን 
# የውሂብ ጎታችን ስም weather_2012 መሆኑን እናውቃለን። 
መጠይቅ = መገናኛ.execute("SELECT * From weather_2012")

# የመጀመሪያ ረድፍ ለ አምዱ ስም እንዲሆን ለመለየት (ኮለምን ስም መለየት)
ስም = [አምድ[0] for አምድ in መጠይቅ.description]

# መገኛውን መጠይቁን በመጠቀም ፓንዳስ ዳታፍሬም መፍጠር 
ውጤት= pd.DataFrame.from_records(data=መጠይቅ.fetchall(), columns=ስም)
```
